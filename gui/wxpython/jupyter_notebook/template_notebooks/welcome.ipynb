## Quick Start

Run the cells below in order to set up your GRASS Jupyter environment:

````
# Import GRASS Jupyter modules
import grass.jupyter as gj

# Initialize GRASS session
gj.init("${MAPSET_PATH}")

print("GRASS Jupyter environment initialized!")
````

````
# Import the GRASS Tools API (recommended)
from grass.tools import Tools

tools = Tools()
print(tools.g_region(flags="p").text)
````

---

## Examples - Try These!

### Example 1: List Available Data

````
# List raster maps in current mapset
tools.g_list(type="raster", mapset=".").text.splitlines()
````

### Example 2: Visualize a Map

*Replace `'elevation'` with a raster from your mapset*

````
# Create a map display
m = gj.Map()

# Add a raster layer (change 'elevation' to your raster name)
m.d_rast(map="elevation")

# Display the map
m.show()
````

### Example 3: Run Analysis with Tools API

````
# Get statistics for a raster map
# Replace 'elevation' with your raster name
tools.r_univar(map="elevation", format="json")
````

### Tip: Tools API vs. Traditional API

**New Tools API (recommended):**
```python
tools.r_slope_aspect(elevation="elevation", slope="slope")
```

**Traditional API:**
```python
import grass.script as gs
gs.run_command("r.slope.aspect", elevation="elevation", slope="slope")
```

The Tools API provides better autocomplete, type hints, and cleaner syntax!
