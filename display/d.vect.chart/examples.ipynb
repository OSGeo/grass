# Graphics for Documentation of _d.vect.chart_

Requires _pngquant_, _optipng_ and _ImageMagic_ (_mogrify_, _montage_).

```python
from IPython.display import Image

import grass.jupyter as gj
```

```python
gj.init("~/grassdata/nc_spm_08_grass7/user1")
```

```python
plot = gj.Map(width=600)
# plot.d_background(color="white")
plot.d_vect(map="streets_wake", color="#d6d6d6", flags="s")
plot.d_vect_chart(
    map="firestations",
    columns="PUMPERS,TANKER,PUMPER_TAN,MINI_PUMPE",
    size_column="PUMPERS+TANKER+PUMPER_TAN+MINI_PUMPE",
    scale=8,
    outline_color="255:255:255:255",
    colors="#00cb85,#537eff,#efe645,#e935a1",
    flags="c",
    legend_title="Firefighting equipment",
)
plot.d_legend_vect(at=[70, 25])
plot.show()
filename = "d_vect_chart_legend.png"
plot.save(filename)
!mogrify -trim {filename}
!pngquant --ext ".png" -f {filename}
!optipng -o7 {filename}
Image(filename)
```
